```python
import os
import time
import datetime
import pandas as pd
path = "C:\\Documenti\\"
```

```python
valuta = 'EURUSD=X'
data_inizio = int(time.mktime(datetime.datetime(2021, 1, 1).timetuple()))
data_fine = int(time.mktime(datetime.datetime(2022, 1, 1).timetuple()))
frequenza = '1d'
```

```python
query = f'https://query1.finance.yahoo.com/v7/finance/download/{valuta}?period1={data_inizio}&period2={data_fine}&interval={frequenza}&events=history&includeAdjustedClose=true'
```

```python
print(query)
```

```python
USD = pd.read_csv(query)
USD.head(20)
```

```python
USD.to_csv("C:/Documenti/USD.csv")
```
